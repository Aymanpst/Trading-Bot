```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from pathlib import Path



```

```python
DATA_DIR = Path('../SCR/data')

sp500 = pd.read_csv(DATA_DIR / 'sp500_m5.csv',  index_col='time', parse_dates=True)
gold  = pd.read_csv(DATA_DIR / 'or_m5.csv',     index_col='time', parse_dates=True)
brent = pd.read_csv(DATA_DIR / 'brent_m5.csv',  index_col='time', parse_dates=True)
eurusd= pd.read_csv(DATA_DIR / 'eurusd_m5.csv', index_col='time', parse_dates=True)

```

```python
import ta

sp500['rsi_5']  = ta.momentum.RSIIndicator(sp500['close'], window=5).rsi()
sp500['rsi_10'] = ta.momentum.RSIIndicator(sp500['close'], window=10).rsi()
sp500['rsi_14'] = ta.momentum.RSIIndicator(sp500['close'], window=14).rsi()

macd = ta.trend.MACD(sp500['close'], window_slow=8, window_fast=3, window_sign=5)
sp500['macd']        = macd.macd()
sp500['macd_signal'] = macd.macd_signal()
sp500['macd_hist']   = macd.macd_diff()

for k in [3, 5, 10]:
    sp500[f'roc_{k}'] = ta.momentum.ROCIndicator(sp500['close'], window=k).roc()

stoch = ta.momentum.StochasticOscillator(sp500['high'], sp500['low'], sp500['close'], window=5, smooth_window=3)
sp500['stoch_k'] = stoch.stoch()
sp500['stoch_d'] = stoch.stoch_signal()

print("✅ Features momentum calculées")
```

```python
sp500['volume_rel'] = sp500['tick_volume'] / sp500['tick_volume'].rolling(20).mean()
sp500['obv']        = ta.volume.OnBalanceVolumeIndicator(sp500['close'], sp500['tick_volume']).on_balance_volume()
sp500['volume_3']   = sp500['tick_volume'].rolling(3).sum()

print("✅ Features volume calculées")
```

```python
sp500['atr_5']  = ta.volatility.AverageTrueRange(sp500['high'], sp500['low'], sp500['close'], window=5).average_true_range()
sp500['atr_14'] = ta.volatility.AverageTrueRange(sp500['high'], sp500['low'], sp500['close'], window=14).average_true_range()

bb = ta.volatility.BollingerBands(sp500['close'], window=20, window_dev=2)
sp500['bb_upper'] = bb.bollinger_hband()
sp500['bb_lower'] = bb.bollinger_lband()
sp500['bb_width'] = bb.bollinger_wband()

sp500['hl_range'] = sp500['high'] - sp500['low']

print("✅ Features volatilité calculées")
```
